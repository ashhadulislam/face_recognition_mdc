## Step 3. Calculate the normalised faces

```python
mean_faces_broad=(matrix_vectors-avg_vector)
mean_faces_tall=mean_faces_broad.transpose()
print(mean_faces_broad.shape)
print(mean_faces_tall.shape)
```

## Step 4. Calculate covariance matrix

```python
cov_matrix=np.cov(mean_faces_broad)
cov_matrix = np.divide(cov_matrix,8.0)
print('Covariance matrix of X: \n%s' %cov_matrix)

```

```python
eigenvalues, eigenvectors, = np.linalg.eig(cov_matrix)
print('Eigenvectors of Cov(X): \n%s' %eigenvectors)
print('\nEigenvalues of Cov(X): \n%s' %eigenvalues)
```

```python
idx = eigenvalues.argsort()[::-1]   
eigenvalues = eigenvalues[idx]

print("Number of eigen values = ",len(idx))

# to sort the eigen vectors
list_of_eigen_vectors=[]
for index in idx:
    list_of_eigen_vectors.append(eigenvectors[index])
eigen_vectors=np.array(list_of_eigen_vectors)
```
